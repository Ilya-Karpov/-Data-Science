При фиксированном seed=21  поэкспериментируйте с параметрами алгоритма дифференциальной эволюции: strategy, popsize,tol, mutation, recombination и updating.

Постройте графики количества итераций оптимизации функции Растригина от значения параметра.

```python
import numpy as np
```

```python
def rastrigin(X):
    A = 10
    return A * len(X) + sum((x**2 - A * np.cos(2 * np.pi * x)) for x in X)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.gca(projection='3d')
ax.view_init(20, 35)

X = np.arange(-5.12, 5.12, 0.1)
Y = np.arange(-5.12, 5.12, 0.1)
X, Y = np.meshgrid(X, Y)
Z = rastrigin(np.array([X, Y]))

surf = ax.plot_surface(X, Y, Z, cmap='nipy_spectral')
plt.show()
```

```python
from scipy.optimize import differential_evolution
```

```python
bounds = [(-10, 10), (-10, 10), (-10, 10)]
```

```python
result = differential_evolution(rastrigin, bounds,  seed = 21, strategy='currenttobest1bin', popsize=10,  recombination = 0.7, tol = 0.0001, mutation=(0.5, 1), updating='immediate')
```

```python
strategy = []
```

```python
result_str = differential_evolution(rastrigin, bounds,  seed = 21, strategy='rand1bin')
strategy.append(result_str.nit)
strategy
```

График количества итераций оптимизации функции Растригина от значения параметра "strategy"
