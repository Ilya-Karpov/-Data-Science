Проведите вычислительный эксперимент - для заданной функции Растригина найдите среднее и минимальное значение целевой функции по результатам 100 экспериментов, а также среднее число итераций при случайных начальных условиях в диапазон [-5;5]. Эксперимент выполнить для размерностей d = 2 и 4.

```python
import numpy as np
```

```python
def rastrigin(X):
    A = 10
    return A + sum((x**2 - A * np.cos(2 * np.pi * x)) for x in X)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.gca(projection='3d')
ax.view_init(20, 35)

X = np.arange(-5.12, 5.12, 0.1)
Y = np.arange(-5.12, 5.12, 0.1)
X, Y = np.meshgrid(X, Y)
Z = rastrigan(np.array([X, Y]))

surf = ax.plot_surface(X, Y, Z, cmap='nipy_spectral')
plt.show()
```

```python
from scipy.optimize import minimize
```

```python
x0 = np.random.uniform(low = -5, high = 5, size =(2,))
```

```python
res = minimize(rastrigin, x0, method = 'nelder-mead', options = {'xtol': 1e-8, 'disp': True})
```

Для размерности d = 2

```python
itList2 = [];
resList2 = [];

for kk in range(100):
    x0 = np.random.uniform(low = -5, high = 5, size =(2,))
    res = minimize(rastrigin, x0, method = 'nelder-mead', options = {'xtol': 1e-8, 'disp': False})
    
    itList2.append(res.nit)
    resList2.append(rastrigin(res.x))
```

```python
np.mean(resList2)
```

```python
np.min(resList2)
```

```python
np.mean(itList2)
```
